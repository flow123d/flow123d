```python
from sympy import Matrix, symbols, eye

def symbolic_inverse(rows, cols):
    # Generate symbolic variables for the matrix
    matrix_elements = [[symbols(f'a{i+1}{j+1}') for j in range(cols)] for i in range(rows)]
    
    # Create the matrix using SymPy
    A = Matrix(matrix_elements)
    
    # Check for square matrix to decide between inverse and pseudoinverse
    if rows == cols:
        # Compute the symbolic inverse
        return A.inv()
    else:
        # Compute the symbolic pseudoinverse
        return (A.T * A).inv() * A.T if rows > cols else A.T * (A * A.T).inv()

pinv23 = symbolic_inverse(2,3)
display(pinv23)

spinv23 = sy.separatevars(pinv23)
#pinv23 = sy.simplify(pinv23)
display(pinv23)
#sum of squares of subdets:
#(a11*a22 - a12*a21)**2 + ...
```
